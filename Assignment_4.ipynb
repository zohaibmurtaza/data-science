```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, manhattan_distances, euclidean_distances
import pandas as pd
import numpy as np
```

```python
# Define the sentences
sentences = [
    "data science is one of the most important courses in computer science",
    "this is one of the best data science courses",
    "the data scientists perform data analysis"
]
```

```python
# Tokenization and BoW
count_vectorizer = CountVectorizer()
bow_matrix = count_vectorizer.fit_transform(sentences)
bow_tokens = count_vectorizer.get_feature_names_out()
df_bow = pd.DataFrame(data=bow_matrix.toarray(), columns=bow_tokens)
df_bow
```

```python
# TF
tf_matrix = np.divide(bow_matrix.toarray(), np.sum(bow_matrix.toarray(), axis=1)[:, np.newaxis])
df_tf = pd.DataFrame(data=tf_matrix, columns=bow_tokens)
df_tf
```

```python
# IDF
idf_matrix = np.log(np.divide(len(sentences), np.sum((bow_matrix > 0).astype(int), axis=0)) + 1)
df_idf = pd.DataFrame(data=idf_matrix, index=['IDF'], columns=bow_tokens)
df_idf
```

```python
# TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
tfidf_tokens = tfidf_vectorizer.get_feature_names_out()
df_tfidf = pd.DataFrame(data=tfidf_matrix.toarray(), columns=tfidf_tokens)
df_tfidf
```

```python
# Similarity
cosine_similarity_matrix = cosine_similarity(tfidf_matrix)
manhattan_distance_matrix = manhattan_distances(tfidf_matrix)
euclidean_distance_matrix = euclidean_distances(tfidf_matrix)

df_cosine_similarity = pd.DataFrame(data=cosine_similarity_matrix)
df_manhattan_distance = pd.DataFrame(data=manhattan_distance_matrix)
df_euclidean_distance = pd.DataFrame(data=euclidean_distance_matrix)
```

```python
# Question: 1 BoW, TF, IDF, and then TF.IDF
print("Bag of Words:")
print(df_bow)
print("\nTF:")
print(df_tf)
print("\nIDF:")
print(df_idf)
print("\nTF.IDF:")
print(df_tfidf)
```

```python
# Question: 2 similarity between S1, S2, and S3 using cosine, manhattan, and euclidean distances.
print("\nCosine Similarity:")
print(df_cosine_similarity)
print("\nManhattan Distance:")
print(df_manhattan_distance)
print("\nEuclidean Distance:")
print(df_euclidean_distance)
```
